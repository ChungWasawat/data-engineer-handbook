```scala
// create classes

case class MatchDetail (
    match_id: String,
    player_gamertag: String,
    previous_spartan_rank: Option[Integer],
    spartan_rank: Option[Integer],
    previous_total_xp: Option[Integer],
    total_xp: Option[Integer],
    previous_csr_tier: Option[Integer],
    previous_csr_designation: Option[Integer],
    previous_csr: Option[Integer],
    previous_csr_percent_to_next_tier: Option[Integer],
    previous_csr_rank: Option[Integer],
    current_csr_tier: Option[Integer],
    current_csr_designation: Option[Integer],
    current_csr: Option[Integer],
    current_csr_percent_to_next_tier: Option[Integer],
    current_csr_rank: Option[Integer],
    player_rank_on_team: Option[Integer],
    player_finished: Option[Boolean],
    player_average_life: Option[String],
    player_total_kills: Option[Integer],
    player_total_headshots: Option[Integer],
    player_total_weapon_damage: Option[Integer],
    player_total_shots_landed: Option[Integer],
    player_total_melee_kills: Option[Integer],
    player_total_melee_damage: Option[Integer],
    player_total_assassinations: Option[Integer],
    player_total_ground_pound_kills: Option[Integer],
    player_total_shoulder_bash_kills: Option[Integer],
    player_total_grenade_damage: Option[Integer],
    player_total_power_weapon_damage: Option[Integer],
    player_total_power_weapon_grabs: Option[Integer],
    player_total_deaths: Option[Integer],
    player_total_assists: Option[Integer],
    player_total_grenade_kills: Option[Integer],
    did_win: Option[Integer],
    team_id: Option[Integer]
)
 
case class Matches (
    match_id: String,
    mapid: Option[String],
    is_team_game: Option[Boolean],
    playlist_id: Option[String],
    game_variant_id: Option[String],
    is_match_over: Option[Boolean],
    completion_date: Option[String],
    match_duration: Option[Integer],
    game_mode: Option[Integer],
    map_variant_id: Option[String]
)

case class MedalsMatchesPlayers (
    match_id: String,
    player_gamertag: String,
    medal_id: String,
    count: Integer
)

case class Medals (
    medal_id: String
    sprite_uri: Option[String],
    sprite_left: Option[Integer],
    sprite_top: Option[Integer],
    sprite_sheet_width: Option[Integer],
    sprite_sheet_height: Option[Integer],
    sprite_width: Option[Integer],
    sprite_height: Option[Integer],
    classification: Option[String],
    description: Option[String],
    name: Option[String],
    difficulty: Option[Integer]
)
```
